```python
import re
import numpy as np
import matplotlib.pyplot as plt
df.head()
##纽约的
df=df[df.country == 'United States']
df=df[df.market == 'New York']
##改设施数量
df['amenities']=df['amenities'].apply(lambda x: len(x.split(",")))

##更改transit
pd.options.display.max_columns = None
df = df.fillna(0)
length = df["transit"].str.len().fillna(0)
df['transit'] = length
##城市
df= df[df.city.map(type) == str]
#zipcode
df= df[df.zipcode != 0 ]
df= df[df.zipcode != '-- default zip code --']
##更改zip
df['zipcode'] = df.zipcode.apply(lambda x: str(re.findall(r'\d+\b',str(x))[0]))
## 修改superhost
df['host_is_superhost']=df['host_is_superhost'].apply(lambda x: str(x))
df['host_is_superhost']=df['host_is_superhost'].replace({'0':'False','f':'False','t':'True'})
df['host_is_superhost'] = df['host_is_superhost'].apply(lambda x: eval(x))
#房主验证
df['host_identity_verified'] = df['host_identity_verified'].apply(lambda x: str(x)).replace({'t':'True','f':'False'}).apply(lambda x: eval(x))
##更改extra people
df['extra_people'] = df['extra_people'] .apply(lambda x: float(x.split('$')[1]))

df['over_score'] = df['review_scores_rating']*df['review_scores_accuracy']
df['over_score'] = df['over_score']/10
```

```python

column_tranformer = ColumnTransformer(
    transformers = [
        ('or',OneHotEncoder(handle_unknown = 'ignore'),categorical),
        
        ('ohh',KBinsDiscretizer(n_bins=500, 
                                             encode='ordinal', 
                                             strategy='quantile'),bool_col),
        ('num',StandardScaler(),numeric)
    ], remainder = 'passthrough'
)
pred = Pipeline([
    ('col_trans', column_tranformer),
    ('classifier1',  XGBRegressor(n_estimators = 45,max_depth= 6))])

model = pred.fit(X_train,y_train)
print(model.score(X_train,y_train))
y_pred = model.predict(X_test)
rms = mean_squared_error(y_test, y_pred, squared=False)
print(rms)
    
```
